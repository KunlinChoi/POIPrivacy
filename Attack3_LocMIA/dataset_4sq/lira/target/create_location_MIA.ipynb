```python
import pandas as pd
import random
df0 = pd.read_csv("../data_ori/0.csv")
df1=pd.read_csv("../data_ori/1.csv")
df0_user = list(df0["user_id"].unique())
df1_user = list(df1["user_id"].unique())
print(len(df0_user),len(df1_user))
shadow_user = list(set(df0_user + df1_user))
print(len(shadow_user))

df_train_victim = pd.read_csv("../attack_target/train.csv")
valid_user = list(df_train_victim["user_id"].unique())
print(len(valid_user))
valid_user = [x for x in valid_user if x not in shadow_user]
print(len(valid_user))
target_user = random.sample(valid_user, 20)
print(target_user)
```

```python
target = []
used_locs = []

for user in target_user:
    df_user = df_train_victim[df_train_victim["user_id"]==user]
    user_loc = list(df_user["POI_id"].unique())
    for loc in user_loc:
        used_locs.append(loc)
        pair = (user,loc)
        target.append(pair)
        
#print(target)
print(len(target))
all_locs = list(df_train_victim["POI_id"].unique())
print(len(all_locs))
used_locs = list(set(used_locs))
traj_locs =  [x for x in all_locs if x not in used_locs]
print(len(traj_locs))

out_locactions = random.sample(traj_locs, len(target))
traj_locs =  [x for x in traj_locs if x not in out_locactions]
print(len(out_locactions))
print(len(traj_locs))
```

```python
in_locs_save=[]
out_locs_save =[]
user_save = []
print(len(target))
for i in range(len(target)):
    user,loc = target[i]
    user_save.append(user)
    in_locs_save.append(loc)
    out_locs_save.append(out_locactions[i])

df_save=pd.DataFrame({"users":user_save,"in_loc":in_locs_save,"out_loc":out_locs_save})
df_save.to_csv("./loc_target.csv")
```

```python
def create_dataset(shadow_df,include_traj,idx):
    target_df = shadow_df[shadow_df["trajectory_id"].isin(include_traj)]
    #target_df = target_df.sort_values(by=['localtime'],ascending=(True))
    target_df.to_csv("../data_loc/"+ str(idx)+".csv")
```

```python
import heapq
in_dict = {}
in_len_dict= {}
in_traj = list(df_in["trajectory_id"].unique())
out_traj = list(df_out["trajectory_id"].unique())
shadow_traj = in_traj + out_traj
exclusive_list= []
for traj in shadow_traj:
    in_dict[traj] =[]
    in_len_dict[traj] =0
    
random.seed(42)    
for i in range(256): #how many shadow model
        #print(i)
        
        #print(len(exclusive_list))
        #target_lst =[x for x in shadow_traj if len(in_dict[traj])<=50]
        #print("a",len(shadow_traj))
        it = list(in_len_dict.items())
        #print(it)
        random.shuffle(it)
        include_traj=dict(heapq.nsmallest(309, dict(it).items(), key=lambda x: x[1]))
        #print(include_traj)
        #include_traj =  [key for key,pair in in_dict_lst[-100:]]
        #print(include_traj)
        for traj in include_traj:
            in_dict[traj].append(i)
            in_len_dict[traj] = len(in_dict[traj])
            #print(in_dict[traj])
    
        create_dataset(shadow_df,include_traj,i)
     
df_in_log = pd.DataFrame()
for traj in list(in_dict.keys()):
    print(len(in_dict[traj]))
    df_in_log[traj]= in_dict[traj]
 
df_in_log.to_csv("../data_loc/traj_log.csv")
```
