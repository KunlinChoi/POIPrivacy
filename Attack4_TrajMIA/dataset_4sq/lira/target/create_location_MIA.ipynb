```python
import pandas as pd
df0 = pd.read_csv("../data/128/0.csv")
df1=pd.read_csv("../data/128/1.csv")
```

```python
print(df0)
```

```python
import numpy as np
df0_user = df0["user_id"].unique()
df1_user = df1["user_id"].unique()
df_ret= pd.DataFrame()
for user in df1_user[:50]:
    if user in df0_user:
        df_u_0 = df0[df0["user_id"]==user]
        df_u_1 = df1[df1["user_id"]==user]
        loc0 = df_u_0["POI_id"].values
        loc1 = df_u_1["POI_id"].values
        locout = [loc for loc in loc1 if loc not in loc0]
        locin = loc0
        """
        if len(locout)<len(loc0):
            locin = np.random.choice(loc0, len(locout), replace=False)
        else:
            locout = np.random.choice(locout, len(loc0), replace=False)
            locin = loc0
        """
        print(locin)
        print(locout)
        df_in_temp=pd.DataFrame()
        for loc in locin:
            df_in = df_u_0[df_u_0["POI_id"]==loc]
            df_in = df_in.sample(n=1)
            df_in_temp = pd.concat([df_in_temp, df_in], ignore_index=True)
        
        df_in_temp = df_in_temp.drop_duplicates(subset='trajectory_id')
        
        
        
        df_out_temp=pd.DataFrame()
        for loc in locout:
            df_out = df_u_1[df_u_1["POI_id"]==loc]
            df_out = df_out.sample(n=1)
            df_out_temp = pd.concat([df_out_temp, df_out], ignore_index=True)
            
        df_out_temp = df_out_temp.drop_duplicates(subset='trajectory_id')
        
        if len(df_out_temp)<len(df_in_temp):
            df_in_temp= df_in_temp.sample(n=len(df_out_temp))
        else:
            df_out_temp= df_out_temp.sample(n=len(df_in_temp))
            
        if len(df_out_temp) > 2:
            df_in_temp= df_in_temp.sample(n=2)
            df_out_temp= df_out_temp.sample(n=2)
        df_in_temp["label"] = [1] * len(df_in_temp)
        df_out_temp["label"] = [0] * len(df_out_temp)
        df_ret = pd.concat([df_ret, df_in_temp], ignore_index=True)
        df_ret = pd.concat([df_ret, df_out_temp], ignore_index=True)
        print(df_ret)
```

```python
df_ret.to_csv("./loc_target.csv")
```

```python
df_ret_no_time = df_ret
df_ret_no_time["norm_in_day_time"] = [0.5] * len(df_ret_no_time)
df_ret_no_time["UTC_time"] = ["2023-05-32 00:12:00+00:00"] * len(df_ret_no_time)
print(df_ret_no_time)
```
